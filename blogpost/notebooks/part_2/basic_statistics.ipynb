## Get data

We'll use Bitcoin exchange rate. This is a famous example but really interesting because there has been a lot of research on this time series because of its volatility. It is barely unpredictable. 

In order to get such data, I chose [quandl](https://www.quandl.com/), a source for financial data. There is a really good [Python library](https://www.quandl.com/tools/python) you can use to avoid downloading files on your local machine. You may need to create an account in order to get an API key. Be careful, do not expose your credentials on the Internet. You can protect easily your API key in a few lines by following the guidelines in the [README file](https://github.com/quandl/quandl-python) of the repo.

```python
quandl.read_key()
bitcoin_ex_rate = quandl.get('BITSTAMP/USD')
```

```python
bitcoin_ex_rate.head()
```

```python
plt.figure(figsize=(15,10))
plt.xlabel('Time', fontsize=15)
plt.ylabel('Bitcoin exchange rate (in USD)', fontsize=15)
plt.grid(True)
plt.plot(bitcoin_ex_rate['Last'])
plt.title('Evolution of the bitcoin exchange rate (BTC/USD) since 2014')
```
